```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib
```

```python
pd.set_option('display.max_columns', 500)
```

```python
#continent
cont = pd.read_csv("countrylists/groupcountries.csv")
ssa = pd.read_csv("countrylists/ssaCountries.csv")
#agdata
ag = pd.read_csv("aglandsqkm.csv").merge(ssa)
#ag.sort_values("_2012", inplace=True, na_position="first")
irr = pd.read_csv("irrigated.csv").merge(ssa)
irr.sort_values("_2012", inplace=True, na_position="first")
tractors = pd.read_csv("tractors.csv").merge(ssa)
tractors.sort_values("_2012", inplace=True, na_position="first")
merchim = pd.read_csv("merchimport.csv").merge(ssa)
merchim.sort_values("_2012", inplace=True, na_position="first")
foodim = pd.read_csv("foodimport.csv").merge(ssa)
#foodim.sort_values("_2012", inplace=True, na_position="first")
expend = pd.read_csv("finalexp.csv").merge(ssa)
expend.sort_values("_2012", inplace=True, na_position="first")
#infdata
elec = pd.read_csv("electricity.csv")
elec.sort_values("_2012", inplace=True, na_position="first")
```
